```python
import os
import numpy as np
import pandas as pd

from scipy import stats
```

```python
DATA_DIR = '/Users/rebekahzhang/data/behavior_data'
EXP = "exp2"
DATA_PATH = os.path.join(DATA_DIR, EXP, "trials_training_filtered2.csv")
OUTPUT_DIR = os.path.join(DATA_DIR, f'{EXP}_clustering')
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
trials = pd.read_csv(DATA_PATH)
```

```python
trials.head()
```

```python
trials.keys()
```

```python
trials_non_miss = trials[trials["miss_trial"] == False].copy()
```

```python
len(trials_non_miss)
```

```python
tw_short_all = trials_non_miss[trials_non_miss['group'] == 's']["time_waited"].dropna()
tw_long_all = trials_non_miss[trials_non_miss['group'] == 'l']["time_waited"].dropna()
t_stat, p_value = stats.ttest_ind(tw_short_all, tw_long_all)
print(f'{len(trials_non_miss)} all trials')
print(f"S: n={len(tw_short_all)}, mean={tw_short_all.mean():.3f}±{tw_short_all.std():.3f}s")
print(f"L: n={len(tw_long_all)}, mean={tw_long_all.mean():.3f}±{tw_long_all.std():.3f}s")
print(f"T-test: t={t_stat:.3f}, p={p_value}\n")
```
