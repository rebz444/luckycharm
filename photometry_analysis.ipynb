```python
import os
import shutil

import pandas as pd
```

```python
sheet_url = "https://docs.google.com/spreadsheets/d/1B8KCnku1vQInKOFBuoLeuDND6CEUA4hT6qQ5zuhqvyU/export?format=csv&gid=0"
photometry_log = pd.read_csv(sheet_url)
```

```python
data_dir = '/Users/rebekahzhang/data/behavior_data'
exp = "exp2"
data_folder = os.path.join(data_dir, exp)
photometry_folder = '/Users/rebekahzhang/data/photometry_data'
os.makedirs(photometry_folder, exist_ok=True)
```

```python
sessions_training = pd.read_csv(os.path.join(data_folder, f'sessions_training_{exp}.csv'), index_col=0)
```

```python
# Merge sessions_training with photometry_log on 'date' and 'mouse' columns to find matching rows
sessions_photometry = pd.merge(
    sessions_training,
    photometry_log[['date', 'mouse']],
    on=['date', 'mouse'],
    how='inner'
)
```

```python
# Iterate over the 'dir' column in sessions_photometry and copy each directory to photometry_folder
for src_dir in sessions_photometry['dir']:
    src_path = os.path.join(data_folder, src_dir)
    dest_path = os.path.join(photometry_folder, os.path.basename(src_dir))
    if os.path.exists(src_path):
        # If destination already exists, remove it first to avoid errors
        if os.path.exists(dest_path):
            shutil.rmtree(dest_path)
        shutil.copytree(src_path, dest_path)
    else:
        print(f"Source directory does not exist: {src_path}")
```
